```python
from transformers import AutoModelForCausalLM, AutoTokenizer, GPTQConfig, TrainingArguments
```

# this is my tokenizer for preprocessing(encoding)

```python
tokenizer=AutoTokenizer.from_pretrained("TheBloke/Mistral-7B-Instruct-v0.1-GPTQ")
```

1. we are loading the pretrain model
2. supervised finetuning(label data)
3. RLHF(DPO,PPO)

```python
tokenizer
```

```python
 tokenizer.eos_token
```

```python
tokenizer.pad_token=tokenizer.eos_token
```

# loading the quantized model(GPTQ)

  GPTQ is a One-Shot Quantization, not a Zero-Shot Quantization,

```python
quantization_config_loading = GPTQConfig(bits=4, use_exllama=True, tokenizer=tokenizer)  #Here bits are the no. of bits to quantized to.
```

Inference is the process of using a pre-trained model to make predictions or generate outputs from input data.

```python
model=AutoModelForCausalLM.from_pretrained( "TheBloke/Mistral-7B-Instruct-v0.1-GPTQ",
                                     quantization_config=quantization_config_loading,
                                      device_map="auto")  #auto means it will automatically detect CPU/GPU.
```

```python
print(model)
```

```python
model.config.use_cache=False
model.config.pretraining_tp=1
model.gradient_checkpointing_enable()
```

# FINE-TUNING using PEFT

```python
from peft import prepare_model_for_kbit_training
```

```python
model = prepare_model_for_kbit_training(model)  #Refined model
```

```python
from peft import LoraConfig
```

```python
peft_config=LoraConfig(
    r=16,lora_alpha=16,lora_dropout=0.05, bias="none", task_type="CAUSAL_LM", target_modules= ["q_proj", "v_proj"]
)
```

```python
from peft import get_peft_model
```

```python
model=get_peft_model(model,peft_config)
```

```python
model.print_trainable_parameters()
```

```python
from trl import SFTTrainer
# trl -> Transfer Reinforcement learning
```

```python
from transformers import TrainingArguments
```

```python
training_arguments = TrainingArguments(
        output_dir="mistral-finetuned-samsum",
        per_device_train_batch_size=8,
        gradient_accumulation_steps=1,
        optim="paged_adamw_32bit",
        learning_rate=2e-4,
        lr_scheduler_type="cosine",
        save_strategy="epoch",
        logging_steps=100,
        num_train_epochs=1,
        max_steps=250,
        fp16=True,
    )
```

```python
trainer=SFTTrainer(
    model=model,
    train_dataset=data,
    peft_config=peft_config,
    dataset_text_field="text",
    args=training_arguments,
    tokenizer=tokenizer,
    packing=False,
    max_seq_length=512

)
```

```python
trainer.train()
```

```python
trainer.push_to_hub()
```
